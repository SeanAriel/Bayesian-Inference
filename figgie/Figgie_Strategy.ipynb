```python
SUITS = 4
PRIOR = 1/SUITS
CARDS = 40
HAND = 10
COMMON = 12
UNCOMMON = 10
REMAINING_CARDS = CARDS - COMMON
THEORETICAL_REMAINING_CARDS = CARDS - UNCOMMON
```

```python
SPADES_OBSERVATIONS = range(0, 11)

def compute_simple_range(suit_observations):
    r = {}
    for observation in suit_observations:
        
        remaining_hand = HAND - observation
    
        # Total possible hands
        total_hands = comb(CARDS, HAND)
    
        # Ways to get [obs] spades if spades is the common suit
        ways_obs_spades_common = comb(COMMON, observation) * comb(REMAINING_CARDS, remaining_hand)
    
        # Ways to get [obs] spades if spades has 10 cards
        ways_obs_spades_10 = comb(HAND, observation) * comb(THEORETICAL_REMAINING_CARDS, remaining_hand)
    
        # Probability calculations
        P_obs_spades_common = ways_obs_spades_common / total_hands
        P_obs_spades_not_common = ways_obs_spades_10 / total_hands
        P_obs_spades = (P_obs_spades_common * PRIOR) + (P_obs_spades_not_common * (1-PRIOR))
    
        # Bayes' Theorem
        P_spades_common_given_obs_spades = (P_obs_spades_common * PRIOR) / P_obs_spades
        r[observation] = round(P_spades_common_given_obs_spades, 2)

    return r 

def plot_simple_range(results):
    x = list(results.keys())
    y = list(results.values())
    plt.figure(figsize=(12, 6))
    plt.plot(x, y, marker='o', linestyle='-')
    for i, value in enumerate(y):
        plt.text(
            x[i], 
            y[i] + 0.035, 
            f'{value:.2f}', 
            ha='center', 
            va='bottom', 
            fontsize=9, 
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2')
        )
    plt.xlabel('Spades Observations In Starting Hand')
    plt.ylabel('Probability of Spades being the Commmon Suit')
    plt.title("Probability of Spades being the Common Suit given Spades' Presence Hand")
    plt.grid(True)
    plt.show()


results = compute_simple_range(SPADES_OBSERVATIONS)
plot_simple_range(results)
```

```python
SPADES_OBSERVATION = 0
HEART_OBSERVATIONS = range(0, HAND - SPADES_OBSERVATION + 1)

def compute_complex_range(suit_observation, other_suit_observations):
    r = {}
    for other_suit_observation in other_suit_observations:
        remaining_suit = HAND - suit_observation - other_suit_observation
    
        # Total number of ways to choose 10 cards out of 40
        total_hands = comb(40, 10)
    
        # Number of ways to get [obs] spades and 5 hearts if spades is the common suit (12 spades)
        ways_obs_spades_5_hearts_common = comb(12, suit_observation) * comb(10, other_suit_observation) * comb(18, remaining_suit)
    
        # Number of ways to get [obs] spades and [obs] hearts if spades has 10 cards
        ways_obs_spades_5_hearts_10 = comb(10, suit_observation) * comb(10, other_suit_observation) * comb(20, remaining_suit)
    
        # Since it's impossible to get [obs] spades out of 8 spades, we don't consider it
        # Probability calculations
        P_obs_spades_5_hearts_common = ways_obs_spades_5_hearts_common / total_hands
        P_obs_spades_5_hearts_not_common = ways_obs_spades_5_hearts_10 / total_hands
        P_obs_spades_5_hearts = (P_obs_spades_5_hearts_common * PRIOR) + (P_obs_spades_5_hearts_not_common * (1-PRIOR))
    
        # Bayes' Theorem
        P_spades_common_given_5_spades_5_hearts = (P_obs_spades_5_hearts_common * PRIOR) / P_obs_spades_5_hearts
        r[other_suit_observation] = round(P_spades_common_given_5_spades_5_hearts, 2)
    
    return r

def plot_complex_range(results):
    x = list(results.keys())
    y = list(results.values())
    plt.figure(figsize=(12, 6))
    plt.plot(x, y, marker='o', linestyle='-')
    for i, value in enumerate(y):
        plt.text(
            x[i], 
            y[i] + 0.005, 
            f'{value:.2f}',
            ha='center', 
            va='bottom', 
            fontsize=9, 
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.2')
        )
    plt.xlabel('Hearts Observations In Starting Hand')
    plt.ylabel(f'Probability of Spades being the Common Suit, given [{SPADES_OBSERVATION}] Spades')
    plt.title("Probability of Spades being the Common Suit given Spades & Heart's Presence in-Hand")
    plt.grid(True)
    plt.show()


results = compute_complex_range(SPADES_OBSERVATION, HEART_OBSERVATIONS)
plot_complex_range(results)
```
